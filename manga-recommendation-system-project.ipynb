```python
# Import necessary libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import re
```

```python
# Read in the dataset and store it as a data frame
df = pd.read_csv('../input/manga-dataset/manga.csv')
```

```python
# Create new column from values of other columns
df['combined_info'] = df['title'] + ' ' + df['genre'] + ' ' + df['rating'].astype(str)
```

```python
# Create a new numbered column to identify managa
df.insert(0, 'manga_id', range(1, 1 + len(df)))
```

```python
# Convert new column string data into a word count matrix
cm = CountVectorizer().fit_transform(df['combined_info'].values.astype('U'))
```

```python
# Get the cosine similarity matrix from the word count matrix
cs = cosine_similarity(cm)
```

```python
# Loop for searching through the dataset from specified title value
while True:
    # Get the title of the manga
    print('Please, enter the title of your manga of choice.\n')
    title = input()
    if  df.title.str.contains(title, case = False).any():
        manga_id = df[df.title.str.contains(title, case = False)]['manga_id'].values[0] # Define the managa_id variable
        chosen_title = df[df.title.str.contains(title, case = False)]['title'].values[0]  # Define the chosen_title variable
        break
    else:
        print(f'You have entered {title}, and this is not currently an applicable choice.\nPlease try again.\n')   
```

```python
# Create an enumerated list based on the reference of the manga_id variable
cs_scores = list(enumerate(cs[manga_id]))
```

```python
# Sort the cosine sim scores and initialize them to start at position 1
# This skips the first value which is the actual selected title itself
sorted_scores = sorted(cs_scores, key = lambda x:x[1], reverse = True)
sorted_scores = sorted_scores[1:]
```

```python
# Loop to print the recommended titles
j = 0
print(f'The top 10 most recommended manga in relation to {chosen_title} are: \n ')
for item in sorted_scores:
    manga_title = df[df.manga_id == item[0]]['title'].values[0]
    print(j + 1, manga_title)
    j += 1
    if j >= 10:
        break
```
